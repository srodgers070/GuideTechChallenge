```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
filename='NewPursuitGuidanceTest.csv'
data = np.loadtxt(filename, skiprows=1, delimiter=',') #Load it in as an array
print(len(data))
plt.figure(1,figsize=(15,10))
plt.subplot(211)
plt.title("%s X vs Y data" %filename)
plt.plot(data[:,0], data[:,1], ls='--', marker="o", label='Missile Positions')
plt.plot(data[:,2], data[:,3], ls = '-', marker='^', label="Target Positions")
plt.legend(loc='best')
plt.xlabel("X Coordinate [m]")
plt.ylabel("Y Coordinate [m]")
plt.subplot(212)
plt.plot(data[:,6],data[:,5], ls = "-", Label='Separation')
plt.legend(loc='best')
plt.ylabel('Separation [m]')
plt.xlabel('Time (s)')
plt.savefig("test")
```
